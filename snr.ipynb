```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Signal-to-Noise Ratio (SNR)
def signal_to_noise_ratio(signal, noise):
    mean_signal = np.mean(signal)
    mean_noise = np.mean(noise)
    std_noise = np.std(noise)
    snr = 20 * np.log10((mean_signal - mean_noise) / std_noise)
    return snr

# Replace these arrays with your actual seismic data from earthquake monitoring stations
# The arrays should contain the seismic signal and background noise for each station
# For demonstration, I'm using random data.
num_stations = 10
time_points = 1000

seismic_data = np.random.rand(num_stations, time_points)  # Replace this with your data
noise_data = np.random.rand(num_stations, time_points)    # Replace this with your data

# Calculate SNR for each station
snr_values = []
for station in range(num_stations):
    snr = signal_to_noise_ratio(seismic_data[station], noise_data[station])
    snr_values.append(snr)

# Visualize SNR values
plt.figure(figsize=(10, 6))
plt.bar(range(1, num_stations+1), snr_values)
plt.xlabel('Station Number')
plt.ylabel('Signal-to-Noise Ratio (SNR) [dB]')
plt.title('SNR for Seismic Data at Different Stations')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Function to calculate Signal-to-Noise Ratio (SNR)
def signal_to_noise_ratio(signal, noise):
    mean_signal = np.mean(signal)
    mean_noise = np.mean(noise)
    std_noise = np.std(noise)
    snr = 20 * np.log10((mean_signal - mean_noise) / std_noise)
    return snr

# Replace 'station1_data.csv', 'station2_data.csv', and 'station3_data.csv' with your actual data filenames
# Make sure the data files contain seismic signal and noise information for the recording period specified.
data_files = ['station1_data.csv', 'station2_data.csv', 'station3_data.csv']

# Load and process seismic data for each station
seismic_data = []
noise_data = []

for data_file in data_files:
    # Assuming data in CSV format with columns 'signal' and 'noise', adjust these column names if needed.
    data = pd.read_csv(data_file)
    seismic_data.append(data['signal'].values)
    noise_data.append(data['noise'].values)

# Calculate SNR for each station
snr_values = []
for station in range(len(data_files)):
    snr = signal_to_noise_ratio(seismic_data[station], noise_data[station])
    snr_values.append(snr)

# Visualize SNR values
stations = ['KAPI', 'COCO', 'WRAB']
plt.figure(figsize=(8, 6))
plt.bar(stations, snr_values)
plt.xlabel('Station')
plt.ylabel('Signal-to-Noise Ratio (SNR) [dB]')
plt.title('SNR for Seismic Data at Different Stations')
plt.grid(True)
plt.show()
```
